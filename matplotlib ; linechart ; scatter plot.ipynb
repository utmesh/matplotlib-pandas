```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10.10) ## starting point, ending point, no. of points in between
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("time")
plt.ylabel("function of time")
plt.title("cool chart")
plt.plot(x,y)
plt.show()
```

# Scatter plot

```python
import pandas as pd
A = pd.read_csv("data_1d.csv",header = None).as_matrix()
x = A[:,0] # all row column 0 as x-axis
y = A[:,1]
plt.scatter(x,y)
plt.show()
```

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1
plt.scatter(x,y)
plt.plot(x_line,y_line)
plt.show()
```
