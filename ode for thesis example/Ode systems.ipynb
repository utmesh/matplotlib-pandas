```python
%matplotlib inline
import math
from math import sin,cos
import matplotlib.pyplot as plt

def elastance(t):
    tp = 0.35 #peak contraction time
    ts = 0.8  #total contraction time
    T = 1.0 # heart period
    t = t % T # implement periodic elastance
    return np.piecewise(t,[t<tp,np.logical_and(tp<=t,t<ts),t>=ts],[lambda x: np.sin(math.pi*x/(2*tp))**2,
                                                                lambda x:np.cos(math.pi*(x-tp)/(2*(ts-tp)))**2,0])
    #numpy.piecewise(x, condlist, funclist,)
    #Given a set of conditions and corresponding functions, 
    #evaluate each function on the input data wherever its condition is true.
    #creation of anonymous functions 
    #(i.e. functions that are not bound to a name) at runtime, using a construct called "lambda"


t=np.linspace(0,2,1000) #numpy.linspace(start, stop, num=50) # only for plotting
f=elastance(t)     #Returns num evenly spaced samples, calculated over the interval [start, stop] 
#elastance()

plt.plot(t,f)
```
