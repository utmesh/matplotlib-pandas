```python
% matplotlib inline
from scipy.integrate import odeint
import math
from math import sin,cos
import matplotlib.pyplot as plt

def elastance(t):
    tp = 0.35
    ts = 0.8
    T = 1.0
    t = t % T
    return np.piecewise(t,[t<tp,np.logical_and(t>=tp,t<ts),t>=ts],[lambda x: np.sin((np.pi*x)/(2*tp))**2,
                                                                lambda x: np.cos((np.pi*(x-tp))/(2*(ts-tp)))**2,0])
t = np.linspace(0,2,1000)
f = elastance(t)
plt.plot(t,f)

```

```python
from scipy.integrate import odeint
def model(y,t):
    k = 0.1
    dydt = -k*y
    return dydt
y0 = 5
t = np.linspace(0,20)
y = odeint(model,y0,t)
plt.plot(t,y)
```
