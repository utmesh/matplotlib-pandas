```python
import numpy as np
import pandas as pd
import xlrd
%matplotlib inline
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


xl = pd.ExcelFile('LV_data_Krell_study.xlsx') # loading spreadsheet
print(xl.sheet_names) # printing sheet name
df1 = xl.parse('Tabelle1') # loading sheet into dataframe
print(df1.shape)
df1.head()
```

```python
sns.kdeplot(df1.EF.dropna(),color="red")
sns.kdeplot(df1.ESV.dropna(),color="green")
sns.kdeplot(df1.Age.dropna(),color="blue")
sns.kdeplot(df1.Height.dropna(),color="orange")
sns.kdeplot(df1.Weight.dropna(),color="black")
sns.kdeplot(df1.EDV.dropna(),color="purple")
sns.kdeplot(df1.SV.dropna(),color="brown")
sns.kdeplot(df1.Herzfrequenz.dropna(),color="yellow")
# Get current size
fig_size = plt.rcParams["figure.figsize"]
print(fig_size) #100, 60
# Set figure width to 12 and height to 9
fig_size[0] = 18
fig_size[1] = 8
plt.show()

```
